```python
# import python libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
```

```python
df.shape
```

```python
df.head(2)
```

```python
df.info()
```

```python
#drop unrelated/blank columns
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
#check for null values
pd.isnull(df).sum()
```

```python
# drop null values
df.dropna(inplace=True)
```

```python
# change data type
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
df.rename(columns= {'Marital_Status':'Shaadi'})
```

```python
df.describe()
```

```python
df[['Age', 'Orders', 'Amount']].describe()
```

**Exploratory Data Analysis**

**Gender**

```python
ax = sns.countplot(x = 'Gender',data = df,  palette={"M": "Orange", "F": "Blue"})
plt.title("Gender vs Count")
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen, palette={"M": "Orange", "F": "Blue"})
plt.title("Gender Vs Amount")
plt.show()
```

*From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men.*

**Age**

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')
plt.title("Age Group")
plt.show()
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
custom_palette = sns.color_palette("husl", len(sales_age))
sns.barplot(x='Age Group', y='Amount', data=sales_age, palette=custom_palette)
plt.title("Age group Vs Amount")
plt.show()
```

*From above graphs we can see that most of the buyers are of age group between 26-35 yrs female.*

**States**

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
colors = sns.color_palette("husl", len(sales_state))

# Create the bar plot with specified colors
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state, x='State', y='Orders', palette=colors)
plt.title("Total Orders by State")
plt.show()
```

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
colors = sns.color_palette("husl", len(sales_state))
sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(data=sales_state, x='State', y='Amount', palette=colors)
plt.title("Sales- Top 10 States")
plt.show()
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively.*

Marital Status

```python
colors = sns.color_palette("husl", len(df['Marital_Status'].unique()))
ax = sns.countplot(data=df, x='Marital_Status', palette=colors)
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Count Vs Marital_status")
plt.show()
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
plt.title("Group by : Marital_Status")
plt.show()
```


From above graphs we can see that most of the buyers are married (women) and they have high purchasing power.

**Occupation**

```python
colors = sns.color_palette("husl", len(df['Occupation'].unique()))
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data=df, x='Occupation', palette=colors)
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Count By Occupation")
plt.show()
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
colors = sns.color_palette("husl", len(sales_state))

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state, x='Occupation', y='Amount', palette=colors)
plt.title("Buyers Occupation")
plt.show()
```

*From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector.*

**Product Category**

```python
colors = sns.color_palette("husl", len(df['Product_Category'].unique()))
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data=df, x='Product_Category', palette=colors)
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Count of Product's Sales")
plt.show()
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
colors = sns.color_palette("husl", len(sales_state))
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state, x='Product_Category', y='Amount', palette=colors)
plt.title("Sales of Product")
plt.show()
```

*From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category.*

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
colors = sns.color_palette("husl", len(sales_state))
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state, x='Product_ID', y='Orders', palette=colors)
plt.title("Top Selling product")
plt.show()
```
